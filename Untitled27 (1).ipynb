```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# ============================
# إعدادات
# ============================

CSV_FILE = "/content/diseases_3_features.csv"          # اسم ملف البيانات
OUTPUT_FILE = "dataset.csv"

COPIES_PER_DISEASE = 20             # عدد النسخ الجديدة لكل مرض

MODEL_NAME = "google/flan-t5-base"

# ============================
# تحميل الموديل
# ============================

print("Loading model...")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

print("Model Loaded.")

# ============================
# قراءة البيانات
# ============================

df = pd.read_csv(CSV_FILE)

required_columns = [
    "Disease Name",
    "Overview",
    "Treatment",
    "Symptoms"
]

for col in required_columns:
    if col not in df.columns:
        raise Exception(f"Column '{col}' not found.")

new_rows = []

# ============================
# دالة إعادة الصياغة
# ============================

def paraphrase(text):

    prompt = f"""
Paraphrase the following medical text.
Keep ALL medical facts exactly the same.
Do not change disease names.
Only rewrite the wording.

{text}
"""

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=512
    ).to(device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=256,
        do_sample=True,
        temperature=0.9,
        top_p=0.95,
        top_k=50
    )

    return tokenizer.decode(
        outputs[0],
        skip_special_tokens=True
    )

# ============================
# إنشاء البيانات الجديدة
# ============================

total = len(df)

for index, row in df.iterrows():

    print(f"{index+1}/{total} : {row['Disease Name']}")

    # الصف الأصلي
    new_rows.append(row.to_dict())

    for i in range(COPIES_PER_DISEASE):

        try:

            overview = paraphrase(row["Overview"])
            symptoms = paraphrase(row["Symptoms"])
            treatment = paraphrase(row["Treatment"])

            new_rows.append({
                "Disease Name": row["Disease Name"],
                "Overview": overview,
                "Treatment": treatment,
                "Symptoms": symptoms
            })

        except Exception as e:
            print(e)

# ============================
# حفظ الملف
# ============================

expanded = pd.DataFrame(new_rows)

expanded.to_csv(
    OUTPUT_FILE,
    index=False,
    encoding="utf-8-sig"
)

print("=" * 40)
print("Finished Successfully")
print("Original Rows :", len(df))
print("New Rows :", len(expanded))
print("Saved As :", OUTPUT_FILE)
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from tqdm import tqdm


# ============================
# إعدادات
# ============================

CSV_FILE = "/content/structured_medical_dataset.csv"

OUTPUT_FILE = "/content/augmented_medical_dataset.csv"

# عدد النسخ الجديدة لكل مرض
COPIES_PER_DISEASE = 5


MODEL_NAME = "google/flan-t5-base"


# ============================
# تحميل الموديل
# ============================

print("Loading model...")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = AutoModelForSeq2SeqLM.from_pretrained(
    MODEL_NAME
)


device = "cuda" if torch.cuda.is_available() else "cpu"

model.to(device)

print("Model Loaded on:", device)



# ============================
# قراءة البيانات
# ============================

df = pd.read_csv(
    CSV_FILE,
    encoding="utf-8-sig"
)


required_columns = [
    "condition",
    "causes",
    "recommendations",
    "symptoms"
]


for col in required_columns:

    if col not in df.columns:
        raise Exception(
            f"Column '{col}' not found"
        )


print("Dataset size:", len(df))



# ============================
# Paraphrase Function
# ============================

def paraphrase(text):

    if pd.isna(text) or len(str(text)) < 10:
        return text


    prompt = f"""
Rewrite the following medical symptoms.

Rules:
- Keep the exact same medical meaning.
- Do not add new symptoms.
- Do not remove symptoms.
- Do not give diagnosis.
- Write it as a patient describing symptoms.

Symptoms:
{text}
"""


    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=256
    ).to(device)



    outputs = model.generate(
        **inputs,
        max_new_tokens=128,
        do_sample=True,
        temperature=0.9,
        top_p=0.95,
        top_k=50,
        num_return_sequences=1
    )


    result = tokenizer.decode(
        outputs[0],
        skip_special_tokens=True
    )


    return result.strip()



# ============================
# Augmentation
# ============================

new_rows = []


for index, row in tqdm(
    df.iterrows(),
    total=len(df)
):


    condition = row["condition"]

    symptoms = row["symptoms"]

    causes = row["causes"]

    recommendations = row["recommendations"]



    # ============================
    # Original Data
    # ============================

    new_rows.append({

        "condition": condition,

        "symptoms": symptoms,

        "causes": causes,

        "recommendations": recommendations,

        "is_augmented": False
    })



    # ============================
    # New Data
    # ============================

    for i in range(COPIES_PER_DISEASE):

        try:

            new_symptoms = paraphrase(
                symptoms
            )


            new_rows.append({

                "condition": condition,

                "symptoms": new_symptoms,

                "causes": causes,

                "recommendations": recommendations,

                "is_augmented": True

            })


        except Exception as e:

            print(
                "Error:",
                condition,
                e
            )



# ============================
# Save
# ============================

expanded = pd.DataFrame(
    new_rows
)


expanded.to_csv(
    OUTPUT_FILE,
    index=False,
    encoding="utf-8-sig"
)



print("="*50)

print(
    "Original Rows:",
    len(df)
)

print(
    "After Augmentation:",
    len(expanded)
)

print(
    "Saved:",
    OUTPUT_FILE
)

print("="*50)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# 1. تحميل الداتا
df = pd.read_csv('augmented_medical_dataset.csv')

# دمج الأعمدة النصية للأنماط الكاملة (مع معالجة القيم الفارغة إن وُجدت)
df['full_text'] = (
    df['symptoms'].fillna('') + " " +
    df['causes'].fillna('') + " " +
    df['recommendations'].fillna('')
)

X_text = df['full_text']
y_labels = df['condition']
```

```python
# ترميز المتغير المستهدف (condition)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y_labels)

# تقسيم البيانات (80% تدريب - 20% اختبار) مع الحفاظ على التوزيع النسبى للحالات (stratify)
X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text, y, test_size=0.2, random_state=42, stratify=y
)

# تحويل النصوص لـ TF-IDF Vectors
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train_text)
X_test_tfidf = vectorizer.transform(X_test_text)
```

```python
# إنشاء الموديل وتدريبه
model = LogisticRegression(max_iter=1000, C=1.0, random_state=42)
model.fit(X_train_tfidf, y_train)
```

```python
# التنبؤ على مجموعة الاختبار
y_pred = model.predict(X_test_tfidf)

# التقييم
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%\n")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

```python
import pandas as pd
import numpy as np
import os
import joblib
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression

# 1. قراءة ملف البيانات
# تأكد إن ملف 'augmented_medical_dataset.csv' مرفوع في Colab (أو عدل المسار لمكانه)
df = pd.read_csv('augmented_medical_dataset.csv')

# 2. تجهيز الأعراض والـ Labels
X_text = df['symptoms'].fillna('')
y_labels = df['condition']

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y_labels)

X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text, y, test_size=0.2, random_state=42, stratify=y
)

# 3. تحويل النص لـ TF-IDF وتدريب الموديل
vectorizer = TfidfVectorizer(max_features=10000, stop_words='english', ngram_range=(1, 2))
X_train_tfidf = vectorizer.fit_transform(X_train_text)

model = LogisticRegression(max_iter=1000, C=2.0, class_weight='balanced', random_state=42)
model.fit(X_train_tfidf, y_train)

# 4. حفظ الملفات في مجلد مشروعك Criativity_Datassinsy على הـ Drive
save_dir = '/content/drive/MyDrive/Criativity_Datassinsy/my_custom_model_folder'
os.makedirs(save_dir, exist_ok=True)

joblib.dump(vectorizer, os.path.join(save_dir, 'tfidf_vectorizer.pkl'))
joblib.dump(model, os.path.join(save_dir, 'logistic_model.pkl'))
joblib.dump(label_encoder, os.path.join(save_dir, 'label_encoder.pkl'))

print(f"🎉 تم حفظ جميع الملفات بنجاح في المسار:\n{save_dir}")
```

```python
import numpy as np
import re

def clean_symptoms_input(text):
    text = text.lower()
    replacements = {
        "sensitivity to light": "photosensitivity",
        "light sensitivity": "photosensitivity",
        "shortness of breath": "breathlessness",
        "chest pain": "chest_pain"
    }
    for k, v in replacements.items():
        text = text.replace(k, v)

    fillers = [r'\bi have\b', r'\bi feel\b', r'\bseverely\b', r'\bsevere\b', r'\bmild\b', r'\ba\b', r'\ban\b', r'\band\b', r'\bto\b']
    for filler in fillers:
        text = re.sub(filler, '', text)

    return " ".join(text.split())

def predict_disease_only(patient_complaint):
    # 1. تنظيف النص
    cleaned_text = clean_symptoms_input(patient_complaint)

    # 2. تحويل النص لفيكتور
    complaint_tfidf = vectorizer.transform([cleaned_text])

    # 3. التوقع واستخراج اسم المرض
    prediction_index = model.predict(complaint_tfidf)[0]
    disease_name = label_encoder.inverse_transform([prediction_index])[0]

    return disease_name

# 🧪 تجربة:
my_complaint = "a high fever, cough, and runny nose"
result = predict_disease_only(my_complaint)

print(result)
```

```python
import gradio as gr
import pandas as pd
import numpy as np
import re

# تحميل ملف البيانات لاستخراج الأسباب والتوصيات
df_medical = pd.read_csv('augmented_medical_dataset.csv')

def clean_symptoms_input(text):
    text = text.lower()
    replacements = {
        "sensitivity to light": "photosensitivity",
        "light sensitivity": "photosensitivity",
        "shortness of breath": "breathlessness",
        "chest pain": "chest_pain"
    }
    for k, v in replacements.items():
        text = text.replace(k, v)

    fillers = [r'\bi have\b', r'\bi feel\b', r'\bseverely\b', r'\bsevere\b', r'\bmild\b', r'\ba\b', r'\ban\b', r'\band\b', r'\bto\b']
    for filler in fillers:
        text = re.sub(filler, '', text)

    return " ".join(text.split())

# دالة التوقع الأساسية
def diagnose_patient(patient_complaint):
    if not patient_complaint.strip():
        return "⚠️ يرجى إدخال الأعراض أولاً.", "غير متوفر", "غير متوفر"

    cleaned_text = clean_symptoms_input(patient_complaint)
    complaint_tfidf = vectorizer.transform([cleaned_text])
    predicted_idx = model.predict(complaint_tfidf)[0]
    disease_name = label_encoder.inverse_transform([predicted_idx])[0]

    # البحث عن التفاصيل في الداتا سيت
    disease_row = df_medical[df_medical['condition'].str.lower() == disease_name.lower()]

    if not disease_row.empty:
        causes = disease_row['causes'].values[0] if 'causes' in disease_row.columns else 'غير متوفرة في البيانات'
        recommendations = disease_row['recommendations'].values[0] if 'recommendations' in disease_row.columns else 'غير متوفرة في البيانات'
    else:
        causes = "لم يتم العثور على أسباب لهذا المرض في الداتا سيت."
        recommendations = "لم يتم العثور على توصيات لهذا المرض في الداتا سيت."

    return disease_name, causes, recommendations

# 🎨 بناء الواجهة الجمالية
with gr.Blocks(theme=gr.themes.Soft(primary_hue="teal", secondary_hue="blue")) as app:

    # الهيدر واسمك
    gr.Markdown(
        """
        # 🩺 AI Medical Condition Predictor
        ### 👨‍💻 Developed by **Abdelrahman Mohamed** | *Data Scientist & ML Engineer*
        ---
        أهلاً بك! أدخل الأعراض التي تشعر بها باللغة الإنجليزية، وسيقوم نموذج الذكاء الاصطناعي بتشخيص الحالة وعرض الأسباب والتوصيات الطبية فوراً.
        """
    )

    with gr.Row():
        # العمود الأيسر: الإدخال
        with gr.Column(scale=1):
            gr.Markdown("### 📝 أدخل الأعراض")
            input_box = gr.Textbox(
                lines=4,
                placeholder="مثال: I have a severe headache, nausea, and sensitivity to light...",
                label="شرح الأعراض (English Symptoms)"
            )
            submit_btn = gr.Button("🔍 تحليل وتشخيص الحالة (Analyze)", variant="primary")

            gr.Examples(
                examples=[
                    ["I have a high fever, cough, and shortness of breath"],
                    ["Chest pain, dizziness, and sweating"],
                    ["Severe headache, nausea, and sensitivity to light"]
                ],
                inputs=input_box,
                label="💡 أمثلة جاهزة للتجربة"
            )

        # العمود الأيمن: عرض النتائج بكروت ملونة
        with gr.Column(scale=1):
            gr.Markdown("### 📊 نتائج التشخيص")
            output_disease = gr.Textbox(label="🏥 التشخيص المتوقع (Predicted Condition)", interactive=False)
            output_causes = gr.Textbox(label="❓ الأسباب المحتملة (Potential Causes)", lines=4, interactive=False)
            output_recs = gr.Textbox(label="💡 التوصيات والنصائح (Recommendations)", lines=4, interactive=False)

    # الربط عند الضغط على الزرار
    submit_btn.click(
        fn=diagnose_patient,
        inputs=input_box,
        outputs=[output_disease, output_causes, output_recs]
    )

    # الفوتر
    gr.Markdown(
        """
        ---
        <center>
        <small>⚠️ <b>تنبيه طبي:</b> هذا التطبيق مخصص لأغراض تعليمية وتطويرية باستخدام الذكاء الاصطناعي، ولا يغني عن الاستشارة الطبية المباشرة.</small>
        </center>
        """
    )

# 🚀 تشغيل الأبلكيشن وتوليد رابط خاريجي (Public Link)
app.launch(share=True)
```
